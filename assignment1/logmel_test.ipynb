```python
import torch
import torchaudio
from melbanks import LogMelFilterBanks
```

```python
signal, sr = torchaudio.load("test_audio.wav", format="wav")
```

```python
sr
```

```python
melspec  = torchaudio.transforms.MelSpectrogram(
    sample_rate=16000,
    n_fft=400,
    hop_length=160,
    n_mels=80,
)(signal)
```

```python
logmelbanks = LogMelFilterBanks()(signal)
```

```python
import matplotlib.pyplot as plt

log_melspec = torch.log(melspec + 1e-6)

plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1)
plt.imshow(log_melspec.squeeze().numpy(), aspect="auto", origin="lower")
plt.title("Torchaudio MelSpectrogram")
plt.colorbar()

plt.subplot(2, 1, 2)
plt.imshow(logmelbanks.squeeze().numpy(), aspect="auto", origin="lower")
plt.title("LogMelFilterBanks")
plt.colorbar()

plt.show()
```
